# PyTorch Tutorial
### Some math functions and plotting, including sine, cosine, and exponential
<br>https://github.com/ostad-ai/PyTorch-Tutorial

```python
import torch
from matplotlib import pyplot as plt
```

PyTorch has functions similar to those of Numpy. Lets try a few of them to display sine functions.
<br> To use torch in pyplot, we have to convert them to numpy arrays.

```python
xs=torch.linspace(-torch.pi,torch.pi,100)
ys=torch.sin(xs)
zs=torch.cos(xs)
plt.plot(xs.numpy(),ys.numpy(),label='sine')
plt.plot(xs.numpy(),zs.numpy(),label='cosine')
plt.title('sine and cosine functions using PyTorch')
plt.legend(); plt.show()
```

Let's define some complex functions:

```python
def fs(x):
    return torch.sin(x)*torch.exp(-x**2)
def fc(x):
    return torch.cos(x)*torch.exp(-x**2)

ys=fs(xs)
zs=fc(xs)
plt.plot(xs.numpy(),ys.numpy(),label='$sin(x)\cdot e^{-x^2}$')
plt.plot(xs.numpy(),zs.numpy(),label='$cos(x)\cdot e^{-x^2}$')
plt.title('Sine and Cosine with exponetial')
plt.legend(); plt.show()
```

Now, we use some reduction operations such as mean and std:

```python
data=torch.randint(-5,5,(18,))
print('integer data, randomly generated: ',data.tolist())
mean,std=data.float().mean().item(),data.float().std().item()
print('mean: ',mean)
print('std: ',std)
uniques,counts=torch.unique(data,return_counts=True)
plt.bar(uniques.tolist(),counts.tolist())
plt.xlabel('elements'); plt.ylabel('frequency')
plt.show()
```
